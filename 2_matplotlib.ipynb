# Matplotlib
Es una librería que permite la creación de figuras y gráficos mediante Python.
- Permite la creación de gráficos y figuras de manera sencilla y eficiente.
- Permite la integración de gráficos y figuras en un cuaderno de Jupyter.


```python
import matplotlib
```

```python
# Módulo que nos proporciona la interfaz con las diferentes opciones
import matplotlib.pyplot as plt
```

```python
# Mostrar los gráficos integrados dentro de jupyter notebook
%matplotlib inline
```

## Representación gráfica de datos
Si a la función de trazado se le da una matriz de datos, la usará como coordenadas en el eje vertical y utilizará el índice de cada punto de datos en el array como la coordenada horizontal.

```python
lista = [1, 2, 5, 7, 8, 3, 1]

plt.plot(lista)
plt.show()
```

```python
# También podemos proporcionar dos matrices: una para el eje horizontal y otra para el eje vertical
# La variable X e Y deben tener el mismo tamaño
lista2 = [1, 4, 6, 7, 8, 9, 10]
plt.plot(lista2, lista) # En eje x tenemos lista 2 y en eje y lista
plt.show()
```

```python
plt.plot([-3, -1, 0, 4, 7], [1, 4, 6, 7, 8])
plt.show()
```

```python
# Ajustamos las longitudes de los ejes
plt.plot([-3, -1, 0, 4, 7], [1, 4, 6, 7, 8])
plt.axis([-4, 8, 0, 10]) # En el eje horizontal desde -4 hasta 8 y en el vertical desde 0 hasta 10
plt.show()
```

```python
# Si queremos pintar una función matemática
import numpy as np
x = np.linspace(-2, 2, 500) # Array con 500 elementos que van de -2 a 2
y = x**2

plt.plot(x, y)
plt.show()
```

```python
# Modificación del estilo de la gráfica añadiendo información
plt.plot(x, y)
plt.title("Square function")
plt.xlabel("x")
plt.ylabel("y = x**2")
plt.grid(True)
plt.show()
```

```python
# Podemos superponer gráficas y cambiar estilo de líneas
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x+1
plt.plot(x, y, 'b--', x, y2, 'g') # La primera función va a tener una línea discontinua azul y la segunda una línea verde
# Si no ponemos nada, cambia de color aleatoriamente cada función
plt.show()
```

```python
# Podemos superponer gráficas y cambiar estilo de líneas
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x+1
plt.plot(x, y, 'b--')
plt.plot(x, y2, 'g') # Se acumulan las funciones y luego las muestra todas
plt.show()
```

```python
# Puedo añadir leyenda
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x+1
plt.plot(x, y, 'b--', label="x**2")
plt.plot(x, y2, 'g', label="x+1") # Se acumulan las funciones y luego las muestra todas
plt.legend(loc="best") # La coloca donde él estima que es mejor
plt.show()
```

```python
# Podemos agregar dos gráficas que no se superpongan, que se organicen en un grid. Se denominan subplots
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x+1

plt.subplot(1, 2, 1) # 1 fila y 2 columnas, la primera gráfica estará en la primera posición
plt.plot(x, y, 'b--')

plt.subplot(1, 2, 2) # La segunda gráfica estará en la segunda casilla
plt.plot(x, y2, 'g') # Se acumulan las funciones y luego las muestra todas
plt.show()
```

```python
# Para que las gráficas no queden tan ajustadas, podemos hacer la figura más grande
import numpy as np

plt.figure(figsize=(14, 6))

x = np.linspace(-2, 2, 500)
y = x**2
y2 = x+1

plt.subplot(1, 2, 1) # 1 fila y 2 columnas, la primera gráfica estará en la primera posición
plt.plot(x, y, 'b--')
plt.xlabel("x1", fontsize=14)
plt.ylabel("y1", fontsize=14)

plt.subplot(1, 2, 2) # La segunda gráfica estará en la segunda casilla
plt.plot(x, y2, 'g') # Se acumulan las funciones y luego las muestra todas
plt.xlabel("x2", fontsize=14)
plt.ylabel("y2", fontsize=14)
plt.show()
```

## Scatter plots

```python
from numpy.random import rand

x, y = rand(2, 100)
```

```python
plt.scatter(x, y)
plt.show()
```

```python
from numpy.random import rand

x, y = rand(2, 100)
x2, y2 = rand(2, 100)

plt.scatter(x, y, c = 'red')
plt.scatter(x2, y2, c = 'blue')

plt.show()
```

## Histogramas

```python
data = [1, 1.1, 1.8, 2, 2.1, 3.2, 3, 3, 3, 3]

plt.subplot(211) #2 filas, 1 columna y la gráfica que voy a pintar va en la primera celda. Permite sin comas

# Vamos a ver con qué frecuencia se repiten los valores dentro de la lista. 
plt.hist(data, bins=15, rwidth=0.8) # bins es el número de columnas rwidth el ancho.
# Según el número de bins, serán los intervalos que se tienen en cuenta
plt.xlabel("Value")
plt.ylabel("Frequency")

data = [1, 1.1, 1.8, 2, 2.1, 3.2, 3, 3, 3, 3]

plt.subplot(212)

plt.hist(data, bins=5, rwidth=0.8)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```
